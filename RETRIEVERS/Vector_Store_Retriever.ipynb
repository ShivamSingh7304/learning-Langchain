```python
from langchain_core.documents import Document
from langchain_community.vectorstores import Chroma 
```

```python
docs = [
    Document(page_content="Cricket is the most popular sport in India."),
    Document(page_content="Football is played by two teams of eleven players."),
    Document(page_content="Python is a versatile programming language used for AI."),
    Document(page_content="The Eiffel Tower is located in Paris, France."),
    Document(page_content="Machine learning enables computers to learn from data."),
    Document(page_content="football is the  second most popular sport in India.")
]
```

```python
pip install -q langchain-huggingface sentence-transformers
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
vector_store = Chroma.from_documents(
    documents=docs,
    embedding=embedding_model,
    persist_directory="./my_collection"
)
```

```python
retriever = vector_store.as_retriever(search_kwargs={"k":1})
```

```python
query = "second most popular sport in india?"

result = retriever.invoke(query)
```
