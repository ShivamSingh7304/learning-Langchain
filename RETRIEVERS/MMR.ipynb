```python
from langchain_core.documents import Document
from langchain_community.vectorstores import Chroma
```

```python
docs = [
    Document(page_content="Football is played between two teams of eleven players on a rectangular field."),
    Document(page_content="Lionel Messi is widely regarded as one of the greatest football players of all time."),
    Document(page_content="Cristiano Ronaldo has won multiple league titles and UEFA Champions League trophies."),
     Document(page_content="Neymar Jr is also widely regarded as one of the greatest football players of all time."),
    Document(page_content="The FIFA World Cup is the most prestigious international football tournament."),
    Document(page_content="The English Premier League is one of the most competitive football leagues in the world."),
    Document(page_content="FC Barcelona is famous for its possession-based tiki-taka style of football."),
    Document(page_content="Real Madrid has won more UEFA Champions League titles than any other club."),
    Document(page_content="A football goalkeeper is the only player allowed to handle the ball inside the penalty area."),
    Document(page_content="Football tactics such as high pressing and counter-attacking can decide the outcome of a match."),
    Document(page_content="Cristiano Ronaldo  is also widely regarded as one of the greatest football players of all time."),
    Document(page_content="Modern football relies on technologies like VAR to assist referees in making accurate decisions.")
]
```

```python
pip install -q langchain-huggingface sentence-transformers
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
model_embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
vector_store = Chroma.from_documents(
    documents=docs,
    embedding=model_embedding,
    persist_directory="./my_collection"
)
```

```python

retriver = vector_store.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 3 , "fetch_k": 4 , "lambda_mult": 0.5}
)
```

```python
query = "best football player"

result = retriver.invoke(query )
```

```python
print(result)
```

```python
for i , doc in enumerate(result):
  print(f"---result{i+1} is:---")
  print(doc.page_content)
```
