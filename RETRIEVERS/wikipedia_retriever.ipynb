# WIKIPEDIA RETRIEVER
##### WikipediaRetriever searches Wikipedia based on a query and returns relevant articles as LangChain Document objects, making it useful for simple RAG applications without needing your own document collection.

```python
from langchain_community.retrievers import WikipediaRetriever
```

```python
retriever = WikipediaRetriever(top_k_results=2 , lang="en")
```

```python
query = "koala"

#getting relevant documents
docs= retriever.invoke(query)
```

```python
for i, docs in enumerate(docs):
  print(f"Document {i+1}:")
  print(docs.page_content)
```
